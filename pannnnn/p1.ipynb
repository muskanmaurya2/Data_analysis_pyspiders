## Concat

```python
pd.concat([df1,df2])     #concat( iterables )-add 2 dataframe          #axis=0(row)
```

```python
df=pd.concat([df1,df2],axis=0)                                     #axis=0(row)
```

```python
df.reset_index()    #to give new index value to the concat table
```

```python
df=df.reset_index(drop=True)             #drop=True is used to remove old index
```

```python
pd.concat([df1,df2],axis=1)                   #axis=1(column)
```
