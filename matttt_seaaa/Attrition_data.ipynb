## apply

```python
def job_clean(row):
    j=row['JOB']
    d=row['Dname']
    if j!='Auditor' and j!='Dairy Person':
        if d in j:
            return j.replace(d,'').strip(',').strip(',s')
        else:
            return j.removeprefix('Exec Assistant').strip(',v')
        

## strip is used to remove the end values
    
    
```

```python
df['JOB']=df.apply(job_clean,axis=1)


df
```

## **Display the count of employess hired in each year**

```python
df['Hiredate']
```

reset_index -> to convert into table(DatafRame)

```python
h_df=df['Hiredate'].dt.year.value_counts().sort_values().reset_index()
h_df
```

```python
h_df.columns=['Year','No_of_hirings']
h_df
```

## **Display the termination details**

filter the termination details

```python
t_df=df.loc[df['Status']=='Terminated','Term_Date'].dt.year.value_counts().sort_values().reset_index()
t_df.columns=['Year','No_of_Terminations']
t_df
```

## Merge both the tables

on: based on which column we are merging

how:it is used for join contion( )

        Why Full Outer join?

        why not inner join?
             some of the year does not have termination, it will some some values 

```python
pd.merge(h_df,t_df,on='Year',how='outer')
```

```python
df1=pd.merge(h_df,t_df,on='Year',how='outer').fillna(0).astype(int)
df1
```

```python
df1['Retention1']=df1['No_of_hirings']-df1['No_of_Terminations']
df1
```

```python
df1['Retention']=df1['Retention1'].cumsum()
df1=df1.drop(columns='Retention1')
df1
```

```python
df.loc[df['Status']=='Active'].shape
```
