```python
def comp_type(row):
    fam=row['Their_family_count']
    pas=row['Number_of_Passager']
    if pas==1:
        return 'Solo'
    elif fam==0 and pas>1:
        return 'Friends'
    else:
        return 'Family'
```

```python
df['Companion_type']=df.apply(comp_type, axis=1)
```

```python
df
```

```python
def Age(row):
    if Age>0 and Age<=2:
        return "Infant"
    elif Age>2 and Age<=16:
        return "Child"
    elif Age>16 and Age<=25:
        return "Teen"
    elif Age>25 and Age<=59:
        return "adults"
    elif Age>59 and Age<=100:
        return "Seniour"
    
```

```python
df['Age_group']=pd.cut(df['Age'],bins=[0,2,16,25,59,100],labels=["Infant","Child","Teen","adults","Seniour"])
```

```python
Royals=['the Countess','Jonkheer', 'Don', 'Dona', 'Master']
Common=[ 'Mrs', 'Mr', 'Ms', 'Mme','Lady', 'Mlle' ,'Sir','Miss']
Millitary=[ 'Major', 'Master', 'Capt', 'Col']
other_professionals=['Dr','Rev']
```

```python
def Social_Status(row):
    c=row['Title']
    if c in Royals:
        return 'Royality'
    elif c in Common:
        return 'Common'
    elif c in Millitary:
        return 'Millitary'
    elif c in other_professionals:
        return 'other_professionals'

    
```

```python
df['Social_Status']=df.apply(Social_Status,axis=1)
```
